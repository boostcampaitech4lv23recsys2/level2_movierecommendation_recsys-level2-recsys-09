```python
import numpy as np
import pandas as pd
```

```python
ease = pd.read_csv('./weight/ease.csv')
bert = pd.read_csv('./weight/bert.csv')
```

```python
ease8 = ease.groupby('user').head(8)
ease8.head(10)
```

```python
ease8['prior'] = [i for i in range(8)]*31360
```

```python
bert['prior'] = [i for i in range(8,18)]*31360
```

```python
result = pd.concat([ease8, bert])
result
```

```python
result = result.drop_duplicates(['user','item'],keep='first')
result
```

```python
result = result.sort_values(['user','prior'])
```

```python
result = result.groupby('user').head(10)
result
```

```python
result[['user','item']].to_csv('./output/ratio_ensemble_ease_bert.csv',index=False)
```
